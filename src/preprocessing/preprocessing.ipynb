```python
import os

def append_string_to_filenames(directory, string_to_append):
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            name, ext = os.path.splitext(filename)
            new_filename = f"{name}{string_to_append}{ext}"
            # new_filename = f"{filename}{string_to_append}"
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))
def delete_matching_string(directory, string_to_delete):
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            name, ext = os.path.splitext(filename)
            new_name = name.replace(string_to_delete, '')
            new_filename = f"{new_name}{ext}"
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))
def replace_matching_string(directory, string_to_replace, new_string):
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            name, ext = os.path.splitext(filename)
            new_name = name.replace(string_to_replace, new_string)
            new_filename = f"{new_name}{ext}"
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))

directory_path = '../../data/raw/'
string_to_replace = 'ID_state'
# string_to_delete = ',max_kick.0e+00'
new_string = 'id_state'
# string_to_append = '.hdf5'
# delete_matching_string(directory_path, string_to_delete)
# replace_matching_string(directory_path, string_to_replace, new_string)
# append_string_to_filenames(directory_path, string_to_append)
```
